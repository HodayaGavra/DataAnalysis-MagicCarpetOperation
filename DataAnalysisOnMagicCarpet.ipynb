```python
df2=pd.read_csv('/Volumes/GoogleDrive/My Drive/MagicCarpet/Olim2.csv')
print(df2.shape)
print(df2.dtypes)
df2.head()
```

```python
result = df1[['index','xx','full_name','gender','age','weight','marital_status']].append(df2[['index','xx','full_name','gender','age','weight','marital_status']])
result.head()
```

```python
result.dtypes
```

```python
result = result.convert_objects(convert_numeric=True)
result.dtypes
```

```python
result.groupby('gender').weight.agg(['count','min','max','mean','median'])
```

```python
result.groupby(['gender','age']).weight.agg(['count','min','max','mean','median'])
```

```python
avg_weight_by_gender_and_age=result[['gender','age','weight']].groupby(['gender','age']).mean()
avg_weight_by_gender_and_age
```
